```python
#creating function to check distribution after imputing ? value with suitable method
def cat_distribution(df,df_new,variable):
    tmp=pd.concat(
                [df[variable].value_counts()/len(df),
                 df_new[variable].value_counts()/len(df_new)],axis=1)
    tmp.columns=['with_nan','without_nan']
    return tmp
```

# function to create diagnostic plot

```python
from scipy import stats as st
```

```python
def plotvariable(df,variable):
    plt.figure(figsize=(16,4))
    #distribution plot
    plt.subplot(1,4,1)
    sns.distplot(df[variable],label="skew: "+str(np.round(df[variable].skew(),2)))
    plt.legend()
    
    #histogram
    plt.subplot(1,4,2)
    plt.hist(df[variable],alpha=0.5)
    
    #boxplot
    plt.subplot(1,4,3)
    sns.boxplot(df[variable])
    
    #q-q plot
    plt.subplot(1,4,4)
    st.probplot(df[variable],plot=plt)
    plt.tight_layout()
    plt.show()
```

# Outlier detection and handling

As we have seen numerical feature are not normally distributes so to handle outlier we Use IQR method. and cap outlier to lower and higher boundries

```python
df_num=df_new.select_dtypes(include='int64')
df_num=df_num.drop('salary',axis=1)
df_num
```

```python
variables=list(df_num.columns)
variables
```

```python
#creating new dataframe for outlier handling
df_capped=df_new[variables].copy()
```

```python
df_capped
```

### creating box plot for each variable to check outliers

```python
df_new.plot(kind='box',figsize=(20,15),layout=(3,3),sharex=False,subplots=True);
```

### Creating a function to calculate upper and lower boundries for each of feature

```python
def iqr_capping(df,features,factor):
    for i in features:
        q1=df[i].quantile(0.25)
        q3=df[i].quantile(0.75)
        
        iqr = q3 - q1
        
        upper_bound=q3+(factor*iqr)
        lower_bound=q1-(factor*iqr)
        df[i]=np.where(df[i]>upper_bound,upper_bound,
                                      np.where(df[i]<lower_bound,lower_bound,
                                               df[i]))
```

```python
iqr_capping(df_capped,['age','fnlwgt','education-num','capital-gain','capital-loss','hours-per-week'],1.5)
```
